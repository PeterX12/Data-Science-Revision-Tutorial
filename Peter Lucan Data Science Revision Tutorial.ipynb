<h3 align="Left">3.2.1 Data Cleaning </h3><a class="anchor" id="3.2.1"></a>
    <p>The work output produced by your Data Sceience work can only be as good as the data you are using [11]. Therefore data has to be cleaned in order to ensure the data is good. As a result Data Cleaning is regarded as one of the most important steps of this process. Data cleaning is the process of replacing and fixing or removing data that is wrong, corrupted, duplicate, incomplete, corrupted or incorrectly formated from within a data set. In essence it is removing data that does not belong in our data set that is being used to achieve the purpose that we have set.

1. Data cleaning is carrried out by first removing duplicate or irrelevant data. That is data that is not needed for your project. 
2. Next the structure is cleaned to fix structural errrors. Structural errrors can be typos, strange naming conventions or even incorrect capitalization. This can cause mislabeled categories or classes. For example "N/A" and "Not Applicable" could both appear in your data set but they should be analyzed in the same calss.
3. Thirdly unwanted outliers must be filtered. Sometimes you may notice that an outlier exists in the data set that is clearly incorrect. When this happens make it should be removed but make sure it is incorrect before dong so. An example of this case would be if your data consisited of the ages of people in a classroom and one of the data suggested that a person is 400 years old. As this is an incorrect outlier it should earase it.
4. Next missing data must be handled. Missing data must be handled as many algorithims will not accept missing values. There are three options when dealing with missing values. The first and most obvious one is to drop the missing values. Doing so will loose the information so this is only done when it is not critical to the project. The second option is to inut your own data based on estimation produced by observing the other data. This is a better option but once again the intergrity of the data is based on the quality of the estimation. The third anf final option is to alter the way the data is used so it may avoid the null values.
5. As the Data Cleaning stage comes to an end one must ask if the data is valid and only proceed when it is indeed so.

During Data Cleaning you may run into "Noisy Data" [11]. Noisy Data is data that has errors and cannot be read by machines as a result. It can be causeed faulty data collection and teh alike. Noisy data can be handled in the following 3 ways:

<b>1.Binning Method:</b> This method works on data that is sorted to smooth it. Data is divided into sections and then different methods are performed on these sections to complete the task. One can replace all teh data in a section by it's average to compete the task if so it is required.

<b>2. Regression:</b> This method works by smoothing data by fitting it to a regression function. The Regression that can be used can be both linear and non linear.

<b>3. Clustering:</b> This method groups similar data into sections called clusters. Outliers may be undetecter or simply just fall outside the cluster.</p>

<h3 align="Left">3.2.4 Data Reduction</h3><a class="anchor" id="3.2.4"></a>
    <p><b>Data Redcution</b> is needed as Data Mining is a technique that deals with hugh quantities of data [9]. This huge amount of data makes analysis of the data very difficult to perform. In order to make it easier Data Reduction techniques are introduced. Data Reduction increases the storage efficiency and as well as this reduces data strogae and analytics costs. The 4 different steps to acieve data refuction are:

<b>1. Data Cube Aggregation</b> is an operation that is applied to data for the construction of the data cube. A data cube is composed of cells, organised by a measure of groups and dimensions. A cell represents the unique logical intersection in a cube of one member from every dmesnion cube.

<b>2. Attribute Subset Selection </b> is a technique used to select and take all the highly relevant attributes and discards the rest. To perform this attribute selection one must use level of significance and p-value of the attribute. The attribute having p-value greater than the significance can be discarded.

<b>3. Numerosity Reduction:</b> this tehcniques enables the stotage of the model of the data rather than the entire data. A good example of this are the Regression Models.

<b>4. Dimensionality Reduction:</b> this technique reduces the size of the data through encoding the data. This can be done in anyway as long as the data is good after reconstruction from compression. Two examples of effective methods of dimensionality reduction are Wavelet transforms and Principal Component Analysis (PCA).</p>

    <p>There are four main types of machine learning algorithims and <b>Supervised Learning</b> is the first one and this references a learning method where the learning process is done by leraning from an example. The <b>Supervised Learning</b> category is then sub divided into 3 further categories. These are <b>1. Classification, 2. Regression</b> and <b>3. Finally <b>Forecasting</b> is used to predict the future based on present and past data.</p>

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sn
#Take in the data from the smsData.csv file.
dataSet = pd.read_csv('smsData.csv')
dataSet
```

<p>As we can see above our data set has a type column and a text column. The type colum is our result column which either outputs spam for a spam message or Ham for a non-spam message. ham is short for non-spam in software development lingo. The text column contains the messages we are testing. As we know machine learning algortihims do not understand text but they do understand numbers. Therefore we will have to once again encode both columns into numbers as they are currently in text.</p>

#### Data Pre Processing

```python
# Check for null values in the data set.
dataSet.isnull().sum()
```

```python
# Naive Bayes only works with numerical values and not Labels so these must be encoded.
# We encode the values in the Sex Column to be 1 or 0 for Male and Female retrospectively.
sexL = LabelEncoder()
dataSet['type'] = typeL.fit_transform(dataSet['type'])
dataSet
```

```python
# Save correctly formatted arrays into y and X for the Naive Bayes Classifier ALgorithim 
y = dataSet.type
X = dataSet.text

# Create the train and test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 101, stratify=y, test_size = 0.20)

# Check the distribution of the training set.
y_train.value_counts(normalize=True)
```

```python
# Check the distribution of the test set.
y_test.value_counts(normalize=True)
```

```python
# Check the shapes of the training sets.
X_train.shape, y_train.shape
```

<p>The text colums must be converted into numerical values. We can do this by using the Count Vectorizer method. This method converts these values into a matrix by taking the most unique words from each row and placing the number of times they appear into the matrix.</p>

```python
# Converting the words in the train rows into vectors 
vec = CountVectorizer()
X_train_count = vec.fit_transform(X_train.values)
X_train_count.toarray()
```

```python
# Create the Navice Bayes function.
theModel = MultinomialNB()

# X_train_count are the messages converted into their bumerical equivelants.
theModel.fit(X_train_count, y_train)
```

```python
# Check the Test Score. (% of how accurate it is).
X_test_count = vec.transform(X_test)
theModel.score(X_test_count, y_test)
```

```python
# Check the Training Score. (% of how accurate it is).
X_train_count = vec.transform(X_train)
theModel.score(X_train_count, y_train)
```
