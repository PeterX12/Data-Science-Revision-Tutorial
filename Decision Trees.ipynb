# Pre Processing

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
dataSet = pd.read_csv('AdData.csv')
dataSet
```

```python
dataSet.drop('User ID', inplace=True, axis=1)
dataSet
```

```python
dataSet = dataSet.rename(columns={'EstimatedSalary':'Salary'})
```

```python
genderL = LabelEncoder()
```

```python
dataSet['Gender'] = genderL.fit_transform(dataSet['Gender'])
dataSet.head(10)
```

```python
dataSet.shape
```

```python
dataSet.isnull().sum()
```

```python
y = dataSet['Purchased']
X = dataSet.drop(['Purchased'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 101, stratify=y, test_size = 0.20)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
dt_model = DecisionTreeClassifier(random_state=10)
```

```python
dt_model.fit(X_train, y_train)
```

```python
dt_model.score(X_train, y_train)
```

```python
dt_model.score(X_test, y_test)
```

```python
dt_model.predict(X_test)
```

```python
dt_model.predict_proba(X_test)
```

```python
tree.plot_tree(dt_model);
```
